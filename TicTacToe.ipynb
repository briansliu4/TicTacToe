```python
import random
#Ask users which symbol they want

choosing = True

while choosing == True:
    user1 = input("Which symbol do you want? Enter X or O: ").upper()
    user2 = ""
    if user1 == "X":
        user2 = "O"
        choosing = False
    elif user1 == "O":
        user2 = "X"
        choosing = False
    else:
        print("The symbol you've entered is invalid. Please try again.")

players = [user1,user2]
    
currentPlayer = random.choice(players)
print("The person who goes first is " + currentPlayer + ".")

no_winner = True

winner = ""

board = [" "," "," "," "," "," "," "," "," "," "]

# Draw out the playing board that users sees
def play_board():
    
    print("{} | {} | {}".format(board[7], board[8], board[9]))  
    print("{} | {} | {}".format(board[4], board[5], board[6]))
    print("{} | {} | {}".format(board[1], board[2], board[3]))
    
# During the game, these things happen: 
def during_game():
    
    #displays intial board
    play_board()
    
    #while loop that runs the game when no winner is achieved
    while no_winner == True:
        
        # player chooses places their symbol
        choose_position(currentPlayer)
        
        # checks if there's a winner or not
        win_condition()
        tie()
        
        # Change to the other player
        change_player()

        if winner == "X":
            print("The winner is player X.")
        elif winner == "O":
            print("The winner is player O.")
        elif winner == "T":
            print("The game is a tie.")       
    
# Function that lets player choose the position to place their symbol
def choose_position(currentPlayer):

    print("It is player " + currentPlayer + "'s turn.")
    position = input("Choose a position between 1-9: ")
    
    vacant = False
    while not vacant: 
        
        # making sure user input is valid 
        while position not in ["1","2","3","4","5","6","7","8","9"]:
            print("The position you've entered is invalid. Please try again.")
            position = input("Enter which position you want to place your symbol. Choose between 1-9: ")
                
        # making sure position on the game board is unoccupied    
        if board[int(position)] == " ":    
            vacant = True  
        else:    
            print("That space is already occupied.")
            position = input("Enter which position you want to place your symbol. Choose between 1-9: ")
        
        
    # fill the board position with current player's symbol
    board[int(position)] = currentPlayer
    
    # display board
    play_board()
    
# Function to change current player
def change_player():      
    global currentPlayer
    if currentPlayer == "X":
        currentPlayer = "O"
    elif currentPlayer == "O":
        currentPlayer = "X"
    
# Function that checks all win conditions
def win_condition(): 
    # How X user wins 
    global winner
    global no_winner
    
    if board[7] == "X" and board[8] == "X" and board[9] == "X":
        winner = "X"
        no_winner = False
    elif board[4] == "X" and board[5] == "X" and board[6] == "X":
        winner = "X"
        no_winner = False
    elif board[1] == "X" and board[2] == "X" and board[3] == "X":
        winner = "X"
        no_winner = False
    elif board[1] == "X" and board[5] == "X" and board[9] == "X":
        winner = "X"
        no_winner = False
    elif board[7] == "X" and board[5] == "X" and board[3] == "X":
        winner = "X"
        no_winner = False
    elif board[7] == "X" and board[4] == "X" and board[1] == "X":
        winner = "X"
        no_winner = False
    elif board[8] == "X" and board[5] == "X" and board[2] == "X":
        winner = "X"
        no_winner = False
    elif board[9] == "X" and board[6] == "X" and board[3] == "X":
        winner = "X"
        no_winner = False

    # How O user wins    
    if board[7] == "O" and board[8] == "O" and board[9] == "O":
        winner = "O"
        no_winner = False
    elif board[4] == "O" and board[5] == "O" and board[6] == "O":
        winner = "O"
        no_winner = False
    elif board[1] == "O" and board[2] == "O" and board[3] == "O":
        winner = "O"
        no_winner = False
    elif board[1] == "O" and board[5] == "O" and board[9] == "O":
        winner = "O"
        no_winner = False
    elif board[7] == "O" and board[5] == "O" and board[3] == "O":
        winner = "O"
        no_winner = False
    elif board[7] == "O" and board[4] == "O" and board[1] == "O":
        winner = "O"
        no_winner = False
    elif board[8] == "O" and board[5] == "O" and board[2] == "O":
        winner = "O"
        no_winner = False
    elif board[9] == "O" and board[6] == "O" and board[3] == "O":
        winner = "O"
        no_winner = False

    return winner
    return no_winner
# Function that checks if the game is a tie
def tie():
    global winner
    global no_winner
    
    if " " not in board[1::]:
        winner = "T"
        no_winner = False
        
    return winner
    return no_winner
#Start the game

during_game()
```
